```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,6,8,9,4])
y = np.array([7,2,6,7,2])
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x)
plt.show()
```

```python
plt.plot(y)
plt.show()
```

```python
plt.plot(x,y, '+')
plt.show()
```

```python
plt.plot(x,y, marker='+')
plt.show()
```

```python
plt.plot(x,y, marker='+', linestyle='-.')
plt.show()
```

```python
plt.plot(x,y, marker='+', ls='--')
plt.show()
```

```python
plt.plot(x,y, marker='+', linestyle='-.', color="#FF203F")
plt.show()
```

```python
plt.plot(x,y, marker='+', linestyle='-.', color="#F7203F", linewidth=2.5)
plt.show()
```

```python
plt.plot(x, marker='+', linestyle='-.', color="#FF2039")
plt.plot(y, marker='*', linestyle='-.', color="purple", linewidth=2.5)
plt.show()
```

```python
plt.plot(x,y, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(x,y, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.title('something about matplotlib')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(x,y, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.title('something about matplotlib', font='Times New Roman', color = 'red', loc= 'left')
plt.xlabel('x', color = 'yellow')
plt.ylabel('y', color = 'green')
plt.show()
```

```python
plt.plot(x,y, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.title('something about matplotlib', font='Times New Roman', color = 'red', loc= 'left', size = 20)
plt.xlabel('x', color = 'cyan')
plt.ylabel('y', color = 'green')
plt.show()
```

```python
styles = {
    'font' : 'Times New Roman',
    'color' : 'red',
    'size' : 20
}
styles
```

```python
plt.plot(x,y, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.title('something about matplotlib', fontdict=styles)
plt.xlabel('x', color = 'cyan')
plt.ylabel('y', color = 'green')
plt.show()
```

```python
plt.plot(x, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.plot(y, marker='*', linestyle='-.', color="green", linewidth=1.5)
plt.title('something about matplotlib', fontdict=styles)
plt.xlabel('x', color = 'cyan')
plt.ylabel('y', color = 'green')
plt.legend(['abc','xyz'])
plt.show()
```

```python
plt.plot(x,y, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.title('something about matplotlib', fontdict=styles)
plt.xlabel('x', color = 'cyan')
plt.ylabel('y', color = 'green')
plt.legend(['abc','xyz'], loc='upper right')
plt.show()
```

```python
plt.plot(x,y, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.title('something about matplotlib')
plt.xlabel('x', color = 'cyan')
plt.ylabel('y', color = 'green')
plt.legend(['abc','xyz'], loc='right', fontsize=15, labelcolor='linecolor')
plt.show()
```

```python
plt.plot(x, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.plot(y, marker='*', linestyle='-.', color="orange", linewidth=1.5)
plt.title('something about matplotlib')
plt.xlabel('x', color = 'cyan')
plt.ylabel('y', color = 'green')
plt.legend(['abc','xyz'], bbox_to_anchor =(0.6, 1.24), ncol=2)
plt.show()
```

```python
plt.plot(x, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.plot(y, marker='*', linestyle='-.', color="orange", linewidth=1.5)
plt.title('something about matplotlib')
plt.xlabel('x', color = 'cyan')
plt.ylabel('y', color = 'green')
plt.legend(['abc','xyz'], bbox_to_anchor =(0.6, 1.24), ncol=2)
plt.figure(figsize=(10,6))
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.subplot(1,2,2)
plt.plot(y, marker='+', linestyle='-.', color="orange", linewidth=1.5)
plt.show()
```
