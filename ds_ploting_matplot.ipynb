```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# creating data points

x = np.array([1,6,8,9,4])
y = np.array([7,2,6,7,2])
```

```python
# Plotting x against y

plt.plot(x,y)
plt.show()
```

```python
# plotting x

plt.plot(x)
plt.show()
```

```python
# plotting y

plt.plot(y)
plt.show()
```

```python
# Plotting x against y with marker '+'

plt.plot(x,y, '+')
plt.show()
```

```python
# Plotting x against y with marker '+' (using marker argument)

plt.plot(x,y, marker='+')
plt.show()
```

```python
# changing linestyle to -.   ,  other linestyle ( - , : )

plt.plot(x,y, marker='+', linestyle='-.')
plt.show()
```

```python
# changing linestyle to --  , using ls argument

plt.plot(x,y, marker='+', ls='--')
plt.show()
```

```python
# changing color of graph , in hex format

plt.plot(x,y, marker='+', linestyle='-.', color="#FF203F")
plt.show()
```

```python
# changing line width

plt.plot(x,y, marker='+', linestyle='-.', color="#F7203F", linewidth=2.5)
plt.show()
```

```python
plt.plot(x, marker='+', linestyle='-.', color="#FF2039")
plt.plot(y, marker='*', linestyle='-.', color="purple", linewidth=2.5)
plt.show()
```

```python
# labeling x and y axis
# plt.xlabel used to label x axis ,  plt.ylabel used to label y axis

plt.plot(x,y, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Adding a title to the plot

plt.plot(x,y, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.title('something about matplotlib')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Adding a title with custom font, color, and location

plt.plot(x,y, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.title('something about matplotlib', font='Times New Roman', color = 'red', loc= 'left')
plt.xlabel('x', color = 'yellow')
plt.ylabel('y', color = 'green')
plt.show()
```

```python
# changing size of title

plt.plot(x,y, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.title('something about matplotlib', font='Times New Roman', color = 'red', loc= 'left', size = 20)
plt.xlabel('x', color = 'cyan')
plt.ylabel('y', color = 'green')
plt.show()
```

```python
# using this variable when to apply same properties to more places
# Defining a dictionary with style settings

styles = {
    'font' : 'Times New Roman',
    'color' : 'red',
    'size' : 20
}
styles
```

```python
# Using the style dictionary to set title properties

plt.plot(x,y, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.title('something about matplotlib', fontdict=styles)
plt.xlabel('x', color = 'cyan')
plt.ylabel('y', color = 'green')
plt.show()
```

```python
# Plotting a legend (describe elements for a particular area of a graph)

plt.plot(x, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.plot(y, marker='*', linestyle='-.', color="green", linewidth=1.5)
plt.title('something about matplotlib', fontdict=styles)
plt.xlabel('x', color = 'cyan')
plt.ylabel('y', color = 'green')
plt.legend(['abc','xyz'])
plt.show()
```

```python
# change location of legend

plt.plot(x,y, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.title('something about matplotlib', fontdict=styles)
plt.xlabel('x', color = 'cyan')
plt.ylabel('y', color = 'green')
plt.legend(['abc','xyz'], loc='upper right')
plt.show()
```

```python
# apply line color to label color and changing size

plt.plot(x,y, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.title('something about matplotlib')
plt.xlabel('x', color = 'cyan')
plt.ylabel('y', color = 'green')
plt.legend(['abc','xyz'], loc='right', fontsize=15, labelcolor='linecolor')
plt.show()
```

```python
# adjust legend position , bbox_to_anchor =(x, y) for positioning , ncol represent in col

plt.plot(x, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.plot(y, marker='*', linestyle='-.', color="orange", linewidth=1.5)
plt.title('something about matplotlib')
plt.xlabel('x', color = 'cyan')
plt.ylabel('y', color = 'green')
plt.legend(['abc','xyz'], bbox_to_anchor =(0.6, 1.24), ncol=2)
plt.show()
```

```python
plt.plot(x, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.plot(y, marker='*', linestyle='-.', color="orange", linewidth=1.5)
plt.title('something about matplotlib')
plt.xlabel('x', color = 'cyan')
plt.ylabel('y', color = 'green')
plt.legend(['abc','xyz'], bbox_to_anchor =(0.6, 1.24), ncol=2)
plt.figure(figsize=(10,6))
plt.show()
```

```python
# Creating subplots and plotting two graphs side by side

plt.subplot(1,2,1)
plt.plot(x, marker='*', linestyle='-.', color="purple", linewidth=1.5)
plt.subplot(1,2,2)
plt.plot(y, marker='+', linestyle='-.', color="orange", linewidth=1.5)
plt.show()
```
