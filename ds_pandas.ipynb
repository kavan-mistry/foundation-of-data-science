# Pandas besics

```python
import pandas as pd
```

```python
# making data in dictionary

data = {
    'name': ['ram', 'charan', 'om', 'sita', 'someone', 'we'],
    'age': [55,95,42,74,20,65],
    'address': ['rampur', 'somnath', 'pune', 'kolkata', 'benglore' , 'kutch'],
    'contact': [1234567890, 4595851685, 1567459852, 7436951478, 4697264859, 9873251469]
}
```

```python
# converting data in dataframe

df = pd.DataFrame(data)
df
```

```python
# .head() is used for viewing top 5 rows

df.head(2)
```

```python
# .tail() is used to view last rows 

df.tail(3)
```

```python
# slice operation to print 2nd to 3rd row

df[2:4]
```

```python
# .shape is used to know dimentions of data, like (row, column)

df.shape
```

```python
# to known the column name and total numbers the .columns was use.

df.columns
```

```python
# exporting data to csv file

df.to_csv('tes.csv')
```

```python
data2 = {
    'some' : [32,9,415,98,32,98],
    'thing' : ['a' , 'd' , 'c' , 'w' , 't', 'b'],
    'diff' : ['y' , 't' , 'i' , 'q' , 'p', 'j'],
    'rent' : [ 2,58,6,8,4,6 ]
}

df2 = pd.DataFrame(data2)
df2
```

```python
# concatinating 2 df, basically it will add columns inside the dataset

df_final = pd.concat((df,df2), axis=1, join='inner')
df_final
```

```python
# .append() add df2 at end of the dataset, it will fill NaN value if column name not matched.

df3 = df.append(df2)
df3
```

```python
# column name should be same for row concatinating.

df_final3 = pd.concat((df,df2), axis=0, join='inner')
df_final3
```
