Box plot for normalised data , box plot is used to represent data in quartiles (25%,50%,75%) here box plot is formed for size , tip and total_bill with respect to its normalised values.

```python
plt.boxplot(df1)
```

Linear plot between day and tip , frequencies are displayed in a linear fashion

```python
x=df["day"]
y=df["tip"]
plt.plot(x,y)
plt.xlabel("day")
plt.ylabel("tip")
```

Scatter plot , day and tip , scatter plot uses dots to represent the values (here tips with respect to day) , we can say that on saturday most amount of tip(s) are given with some being close to 10 i.e higher monitary value , saturday also has tips spread across value more than other days.

```python

plt.xlabel("Day",color='red')
plt.ylabel("Tip",color="magenta")
plt.scatter(x,y)
```

Bar plot represents values in rectangular bar with height representing the tip and width representing which day it is , saturday has most tips given , thursday and sunday almost equivalent , while friday being the least

```python
plt.bar(x, y, color ='green')
```

Here lies the scatter plot between total bill and tip ,by this scatter plot we can determine the relation between total bill and tips , the value of tips increase with increasing bills but the frequency of tips decrease , while lower the bill more the tip but less money given as tip
