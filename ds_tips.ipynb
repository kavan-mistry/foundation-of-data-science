## Boxplot

A box plot, also known as a box-and-whisker plot, is a graphical representation used to visualize the distribution of a dataset's values and identify potential outliers. The plot consists of a rectangular box that represents the interquartile range (IQR), the middle 50% of the data. The median (50th percentile) is shown as a horizontal line within the box. The "whiskers" extend from the box to the minimum and maximum values within a certain range (often calculated as 1.5 times the IQR). Data points outside this range are marked as individual points, indicating possible outliers. Box plots are effective for summarizing the central tendency, spread, and skewness of a dataset, making them useful for comparing distributions across different categories.

```python
boxplt = df.boxplot(figsize = (10,5))
boxplt
```

# Scatter plot

The scatter plot illustrates the relationship between "Total Bill" and "Tip" amounts from the dataset. Each data point is represented as a dot on the plot, with "Total Bill" values on the x-axis and corresponding "Tip" values on the y-axis. The scatter plot helps visually assess any potential patterns or correlations between the two variables.

```python
x = np.array(df['total_bill'])
y = np.array(df['tip'])

plt.scatter(x, y,color = '#88c999')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

# Line chart

The line chart visualizes the relationship between "Total Bill" and "Tip". Each data point is represented as a marker along a line that connects them. The x-axis displays "Total Bill" values, while the y-axis displays corresponding "Tip" values. The line connecting the markers helps in perceiving trends and tendencies between the two variables.

```python
plt.plot(df['total_bill'], df['tip'], marker='o', linestyle='-', color='#C4AEAD')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.grid(True)
plt.show()
```

## Histogram

A histogram is used to illustrate the distribution of a dataset by dividing it into "bins" and representing the frequency or count of data points that fall within each bin. The x-axis represents the "Total Bill" values, while the y-axis displays the frequency or count of occurrences within each bin.

```python
plt.figure(figsize=(15, 5))

# Histogram for total_bill
plt.subplot(1, 2, 1)
plt.hist(df['total_bill'], bins=20, color='#d8bfd8')
plt.title('Total Bill Distribution')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')

# Histogram for tip
plt.subplot(1, 2, 2)
plt.hist(df['tip'], bins=20, color='#b0c4de')
plt.title('Tip Distribution')
plt.xlabel('Tip')
plt.ylabel('Frequency')

plt.show()
```

# Bar plot

A bar chart is a visual representation that displays categorical data as rectangular bars. It's used to compare the quantities of different categories. The x-axis represents the categorical variable Sex, while the y-axis shows the corresponding counts.

```python
sex_counts = df["sex"].value_counts()

plt.bar(sex_counts.index, sex_counts.values, color='#7ac5cd')
plt.title('Count of Male and Female Customers')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

 # Pie chart

 A pie chart is a circular graphical representation used to display the proportion of different categories within a whole dataset. Each "slice" of the pie represents a category (in this case, different days). The size of each slice corresponds to the proportion of that category within the whole.

```python
day_proportions = df["day"].value_counts(normalize=True)
colors = ['#ff9999','#66b3ff','#99ff99','#c4aead']

plt.pie(day_proportions, labels=day_proportions.index, autopct='%.1f%%', colors=colors)
plt.title('Proportion of Meals on Different Days')
plt.show()
```
